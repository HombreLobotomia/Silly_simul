```python
def randomer(stats, tot):
    shots = np.random.rand(tot, len(stats))
    rounds = shots <= stats
    return  rounds
```

```python
def cleaner(rounds):
    # fondamentalmente: se la prima e' in campo, non puo' aver giocato la seconda
    rounds[:,2] = np.where((rounds[:,0]==True), False, rounds[:,2])
    rounds[:,5] = np.where((rounds[:,3]==True), False, rounds[:,5])
```

```python
def counter(rounds):
    prob = len(rounds[((rounds[:,0]==True)&(rounds[:,1]==True)) | (rounds[:,2]==True) | 
                ((rounds[:,3]==True)&(rounds[:,4]==True)) | (rounds[:,5]==True)])
    return prob
```

```python
tot = 10_000_000
```

```python
second_as_second = np.array([0.62, 0.79, 0.51]*2)
second_as_first = np.array([0.62, 0.79, 0.62*0.79]*2)
```

```python
rounds1 = randomer(second_as_second, tot)
cleaner(rounds1)
print(f'{counter(rounds1)*100/tot:.2f}%')
```

```python
rounds2 = randomer(second_as_first, tot)
cleaner(rounds2)
print(f'{counter(rounds2)*100/tot:.2f}%')
```
